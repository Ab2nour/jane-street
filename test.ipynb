```python
from square_logic import LineLogic
from lines import line_8_zones
from black_squares import black_gen, nb_to_gen_from_black_config
import tqdm
```

```python
all_black_cells = black_gen()
```

```python
all_vals = []
for black_config in all_black_cells:
    line = LineLogic()
    line_8_zones(line)
    line.add_palindrome_logic()
    line.black_logic_destructor(black_config, verbose=False)
    elts_to_gen = nb_to_gen_from_black_config(black_config)
    vals = []
    for elt in tqdm.tqdm(elts_to_gen):
        print(elt)
        all_val = list(line.create_generator(*elt))
        vals.append(all_val)
    all_vals.append(vals)

all_k_to_keep = []
for k in range(len(all_black_cells)):
    to_keep=True
    for elt in all_vals:
        if len(elt)==0:
            to_keep=False
            break
    if to_keep:
        all_k_to_keep.append(k)
for k in all_k_to_keep:
    print(all_black_cells[k])
    print(all_vals[k])
print(len(all_k_to_keep))
        
        
```
