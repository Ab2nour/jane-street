```python
def get_lengths(black_boxes: list[int]) -> list[int]:
    """Renvoie la longueur de chaque séquence de nombre, en fonction des cases noires

    >>> get_lengths([])
    [11]
    >>> get_lengths([0])
    [10]
    >>> get_lengths([2])
    [2, 8]
    >>> get_lengths([3])
    [3, 7]
    >>> get_lengths([0, 10])
    [9]
    >>> get_lengths([2, 10])
    [2, 7]
    >>> get_lengths([2, 6])
    [2, 3, 4]
    """
    lengths = []

    if not black_boxes:
        return [N]

    count = 0
    for i in range(N):
        if i in black_boxes:
            if count != 0:
                lengths.append(count)
            count = 0
        else:
            count += 1

    if count != 0:
        lengths.append(count)

    return lengths
```

```python
def first_digit_is_in_1379(n: int) -> bool:
    """Returns True if first digit of n is among 1, 3, 7 or 9.

    >>> first_digit_is_in_1379(1)
    True
    >>> first_digit_is_in_1379(3)
    True
    >>> first_digit_is_in_1379(7)
    True
    >>> first_digit_is_in_1379(9)
    True
    >>> first_digit_is_in_1379(1379)
    True
    >>> first_digit_is_in_1379(12468)
    True
    >>> first_digit_is_in_1379(72468)
    True
    >>> first_digit_is_in_1379(2)
    False
    >>> first_digit_is_in_1379(204568)
    False
    >>>"""
    n_str = str(n)

    if n_str[0] in ("1", "3", "7", "9"):
        return True
    else:
        return False
```
